```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
# Data Preprocessing
dataset=pd.read_csv('E:/4th sem/4. ma-II/3rd unit/MA-Winston Wayne-Chapter 29 Excel files/work.csv')
```

```python
dataset.describe().transpose()
```

```python
dataset
```

```python
transactions=[]
for i in range(0,1000):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
rule_list = apriori(transactions, min_support = 0.02, min_confidence = 0.95, min_lift = 2, min_length = 2)
```

```python
# Visualizing the list of rules
results = list(rule_list)
for i in results:
    print('\n')
    print(i)
    print('**********')
```
